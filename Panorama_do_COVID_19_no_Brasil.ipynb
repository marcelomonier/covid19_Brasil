## Análise Exploratória dos Dados

A fim de entendermos melhor o *dataset* em comento, alguns comandos básicos serão utilizados, tais como:

O código abaixo nos permite:



*   Importar as bibliotecas necessárias para a manipulação do dataset;

```
# importar as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataFrame
DATA = 'https://raw.githubusercontent.com/marcelomonier/covid19_Brasil/main/owid-covid-data.csv'
df = pd.read_csv(DATA)
df.shape
df.head()
```

Uma parte importante, é descobrir o tamanho exato do dataset, pois isso nos permite, com exatidao, saber a quantidadede dados que será processada.

```
# Verificar as dimensões
print('Dimensões do Dataset',
      '\n Veriáveis (Colunas): ', df.shape[1],
      '\n',
      'Entradas: ', df.shape[0])
```

Como visto na saída acima, o dataset em comento, possui 67 colunas e 166789 linhas.

Outro detalhe que merece ser analisado, é o tipo das variáveis. O conhecimento da forma dos dados torna-se extremamente necessário, pois a partir disso pode-se manipular corretamente cada variável.

```
#Verficação do tipo das variáveis
df.dtypes
```

É importante ressaltar que na análise dos dados não pode faltar a verificação dos dados ausentes. O foco não é na falta e sim o porque não estão presentes. Sendo assim, temos que verificar a causa da ausência dos mesmos.

```
round((df.isnull().sum() / df.shape[0]).sort_values(ascending=False),2)
```

Como é possível verificar, a coluna `date` está no formato object. Sendo assim, é necessário transformar a mesma para o tipo datetime.

```
# Transformar a coluna date para o formato datetime
df['date'] = pd.to_datetime(df['date'])
```

A próxima etapa é substituir o valores nulos de duas colunas (`new_deaths` e `total_deaths`) por **0.0**

```
df['new_deaths'].fillna(0, inplace=True)
df['total_deaths'].fillna(0, inplace=True)
```

Em seguida, será a análisada a data mais recente do dataset. Com essa informação, vamos verificar o país líder em casos nesse momento.

```
#Data mais recente do dataset
df['date'].max()
```

```
# Total de cados

df_1 = df.loc[df['date'] == '2022-03-07'].sort_values(by='total_cases', ascending=False)

df_1.head(20)
```

É possível perceber que os países líderes em casos até a presente data, ou seja, 07/03/2022 são:



1.   **Estados Unidos** (Maior que a América do Sul inteira): 79.339.202 casos
2.   **Índia**: 42.971.308 casos
3.   **Brasil**: 29.077.831 casos
4.   **Fança**: 23.124.590 casos
5.   **Reino Unido**: 19.298.531 casos


Outro dado bem relevante, é o total de mortes por COVID-19.



```
# Total de mortes
df_2 = df.loc[df.date == '2022-03-07'].sort_values(by='total_deaths', ascending=False)
df_2.head(20)
```

Em relação as mortes, o resultado acima nos mostra que os países que lideram, são:



1.   **Estados Unidos**:  960.311 mortes
2.   **Brasil**: 652.648 mortes
3.   **Índia**:  515.210 mortes
4.   **Rússia**: 349.853 mortes
5.   **México**: 319.901 mortes

A fim de melhorar a visualização dos dados, podemos verificar esses casos através de um gráfico de barras.





```
#Gráfico de barras para total de mortes por COVID-19

country_top_5_total_deaths = df.loc[df['date'] == '2022-03-07', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[9:14]

fig, ax = plt.subplots(figsize=(15, 10))

ax.bar('location', 'total_deaths', data=country_top_5_total_deaths)

plt.tight_layout()
```

# Progressão do vírus

Atualmente, segundo as princípais mídias de informação, a progressão do vírus vem perdendo força. Outro fator importante, é que a vacina foi essencial, nos novos casos, para a diminuição de sintomas e/ou enfraquecimento dos mesmos.

Para visualizarmos melhor, será plotado um gráfico de linha.

```
#Gráfico de linhas para mortes no mundo

df.loc[df['location'] == 'World', 'total_deaths'].reset_index(drop=True).plot(figsize=(10,8))
```

### Análise Exploratória para o Brasil

A partir deste momento, com o conhecimento relacionados aos acontecimentos do mundo que conseguimos, podemos nos concentrar no Brasil. Sendo assim, será criado um *dataframe* somente com os casos do nosso país.

```
#Dataframe do Brasil

df_brasil = df.loc[df['location'] == 'Brazil'].copy()

df_brasil.head()
```

```
#Tamanho do dataframe
df_brasil.shape
```

```
#Data mais recente
df_brasil.date.max()
```

```
#Data incial
df_brasil.date.min()
```

Como a saída acima nos mostra, o *dataset* possui iformações entre as datas **26/02/2020** e **07/03/2022**. Portanto, é possível vefrificar o dia do primeiro caso e a primeira morte no Brasil.



```
#Identificar quando foi registrada o primeiro caso no Brasil
df_brasil.loc[df_brasil['total_cases'] == 1.0].head(1)
```

```
#Identificar quando foi registrada a primeira morte no Brasil
df_brasil.loc[df_brasil['total_deaths'] != 0.0].head(1)
```

Segundo o *dataset* em comento, o resultado indica que:


*   O primeiro caso registrado no Brasil, foi no dia **26/02/2020**.
*   A primeira morte registrada, foi no dia **17/03/2020**.

Entretanto, [a reportagem da CNN](https://www.cnnbrasil.com.br/saude/primeira-morte-por-covid-19-no-pais-ocorreu-em-12-de-marco-em-sp-diz-ministerio/), afirma que a primeira morte ocorreu no dia **12 de março** em São Paulo, ou seja, contraria o *dataset* em comento. 

Em nota, o ministério afirma ainda que a segunda morte por covid-19 aconteceu no dia **15 de março**, e no **dia 16**, foram mais três óbitos, em São Paulo. No **dia 17**, foram duas mortes em São Paulo e duas no Rio de Janeiro.

O Ministério afirma que o próximo Boletim Epidemiológico trará a correção das informações, e que as mortes foram incluídas no Sistema de Vigilância Epidemiológica da Gripe (SIVEP-Gripe) ao longo do tempo.





```
#Identificar quantos dias demorou para registrar a primeira morte, considerando o primeiro caso de COVID-19 registrado.
from datetime import date

data_1 = df_brasil.date.loc[21387] - df_brasil.date.loc[21367]

data_1.days
```

A importação da biblioteca `date` nos permite calcular a diferença em dias entre as datas do primeiro caso registrado e a primeira morte. 


Portanto, levando em consideração o *dataset*, a diferença é de 20 dias. 

Se levarmos em consideração **a notícia da CNN**, essa diferença cai para **15 dias**.

Agora, será plotado gráficos de linha a fim de compararmos a evolução dos casos e mortes no Brasil.

```
x = list(range(741))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))

ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala Logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)

plt.tight_layout()
```
